```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
M = float(input("Enter value  "))
x = np.linspace(-10,10,100)
y1 = M * x**2
y2 = M * np.sin(x)
```

```python
plt.figure(figsize=(6,6))
plt.plot(x,y1,label = 'M*x^2',color = 'green',linestyle = '-')
plt.plot(x,y2,label = 'M*sin(x)',color = 'red',linestyle = '--')
plt.legend()
plt.grid(True)
plt.title('Plot of y = M*x^2 and y = M*sin(x)')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
data = {
    'Subjects' : ['DBMS','CC','DAA','PNS','AI'],
    'Scores' : [100,98,99,97,95]
}
df = pd.DataFrame(data)
plt.figure(figsize = (10,6))
colors = sns.color_palette('husl',len(df))
sns.barplot(x='Subjects',y='Scores',data=df,hue='Subjects',palette=colors,legend=False)

for i,score in enumerate(df['Scores']):
    plt.text(i,score+1,str(score),ha='center',fontsize=14)

plt.grid(axis='y',linestyle='--',alpha=0.6)
plt.title('Score Sheet of different Subjects')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.show()
```

```python
Roll = 102317131
np.random.seed(Roll)

data = np.random.randn(50)

fig,axes = plt.subplots(2,2,figsize=(10,10))

axes[0,0].plot(np.cumsum(data),color='red',linestyle='-',marker='o',markersize=4)
axes[0,0].set_title('Cumulative Sum Plot')
axes[0,0].set_xlabel('Index')
axes[0,0].set_ylabel('Cumulative Sum')
axes[0,0].grid(True)

axes[0,1].scatter(range(50),data,color='blue',alpha=0.6)
axes[0,1].set_title('Scatter Plot with Random noise')
axes[0,1].set_xlabel('Index')
axes[0,1].set_ylabel('Value')
axes[0,1].grid(True)

axes[1,0].axis('off')
axes[1,1].axis('off')

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


url = "https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
df = pd.read_csv(url)


plt.figure(figsize=(8, 5))
plt.plot(df['month_number'], df['total_profit'], marker='o', color='blue', linestyle='-')
plt.title("Total Profit Over Months")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for column in df.columns[1:-1]:
    plt.plot(df['month_number'], df[column], marker='o', linestyle='-', label=column)

plt.title("Monthly Sales Data of Different Products")
plt.xlabel("Month Number")
plt.ylabel("Units Sold")
plt.legend()
plt.grid(True)
plt.show()
```

```python
df.plot(kind='bar', figsize=(12, 6))
plt.title("Bar Chart for All Features")
plt.xlabel("Month Number")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```
